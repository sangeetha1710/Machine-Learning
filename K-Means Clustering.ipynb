```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student = pd.read_csv('student_clustering.csv')
```

```python
student.head()
```

```python
sns.scatterplot(x = student['cgpa'], y = student['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters= i)
    km.fit_predict(student)
    wcss.append(km.inertia_)
    
```

```python
wcss
```

```python
sns.lineplot(x = range(1,11), y = wcss)
```

```python
df = student.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(student)
```

```python
y_means
```

```python
df[y_means==2,1]
```
